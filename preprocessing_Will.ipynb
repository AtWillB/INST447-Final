```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
co2_df = pd.read_csv("./data/GCB2022v27_MtCO2_flat.csv")
co2_df.info()
```

```python
co2_df.head()
```

```python
# Dropping all rows that have all of the following columsn as NaN
co2_df.dropna(how="all", subset=['Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Other'], inplace=True)
```

```python
# fixng ugly name
co2_df.rename({"ISO 3166-1 alpha-3":"CountryCode", "Total":"Total_MTCO2"}, axis=1, inplace=True)
```

```python
# Setting the country and year as the index
co2_df.set_index(['Country','Year'], inplace=True)
```

```python
co2_df.to_csv("./data/processed_data/co2_country_year.csv")
```

```python
gdp_df = pd.read_excel("./data/WorldEcoArchive.xls")
```

```python
# remove first row
gdp_df = gdp_df.iloc[1:]
```

```python
gdp_df.rename({'GDP per capita, current prices\n (U.S. dollars per capita)':"Country"}, axis=1, inplace=True)
gdp_df.head()
```

```python
#melting values to get same format as co2_country_year
gdp_df = gdp_df.melt(id_vars='Country', var_name='Year', value_name='GDP_per_capita')
```

```python
# setting index like co2
gdp_df.set_index(['Country', 'Year'], inplace=True)
```

```python
#making sure there are ints or nulls
gdp_df.replace("no data", np.NaN, inplace=True)
```
